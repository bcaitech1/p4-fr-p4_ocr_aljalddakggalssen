```python
from torch.utils.data import DataLoader, Dataset
```

```python
class LatexDataset(Dataset):
    def __init__(self, data, tokenizer):
        self.tokenizer = tokenizer
        
        tmp = []
        for text in data:
            tmp.append(self.tokenizer.encode(text))
            
        self.data = tmp
        self.length = len(self.data)
    
    def __getitem__(self, idx):
        return self.data[idx]
    
    def __len__(self):
        return self.length
```

```python
def pad_stuff(stuff, pad_id, max_len=None):
    len_stuff = [len(a) for a in stuff]
    if max_len is None:
        max_len = max(len_stuff)
        
    tmp = []
    for i, x in enumerate(stuff):
        if len_stuff[i] < max_len:
            gap = max_len - len_stuff[i]
            x += gap * [pad_id]
        if len_stuff[i] > max_len:
            x = x[:max_len]
        tmp.append(x)
        
    return tmp

class CustomCollater:
    def __init__(self, tokenizer, max_len=None):
        self.tokenizer = tokenizer
        self.pad_id = self.tokenizer.padding['pad_id']
        self.max_len = max_len
        
    def custom_collate(self, batch):
        input_ids = torch.LongTensor(pad_stuff([b.ids for b in batch], self.pad_id, self.max_len))
        attn_masks = input_ids.ne(self.pad_id).to(torch.long)
        return input_ids, attn_masks
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
from transformers import RobertaConfig, RobertaForMaskedLM
```

```python
bos_token_id, eos_token_id = tokenizer.encode('').ids
pad_token_id = tokenizer.padding['pad_id']
```

```python
config = RobertaConfig(
    pad_token_id=pad_token_id,
    bos_token_id=bos_token_id,
    eos_token_id=eos_token_id,
    hidden_size=192,
    num_attention_heads=4,
    num_hidden_layers=4,
    vocab_size=tokenizer.get_vocab_size(),
)

model = RobertaForMaskedLM(config)
model.to(device)
print()
```

```python
from torch.optim import AdamW
from transformers import get_linear_schedule_with_warmup

epochs = 40
warmup_ratio = 0.1
lr = 1e-4
input_max_len = 250

use_amp = True
max_grad_norm = 1

log_step = 50
save_epoch = 10 # epochs
result_dir = '/opt/ml/code/results/pretrain_latex'
batch_size = 128
model_save_name = 'only_extra'

use_wandb = True

collater = CustomCollater(tokenizer, None)
dl = DataLoader(ds, batch_size=batch_size, collate_fn=collater.custom_collate, num_workers=4, 
                    drop_last=True, shuffle=True)
val_dl = DataLoader(val_ds, batch_size=batch_size, collate_fn=collater.custom_collate, num_workers=4)

optimizer = AdamW(model.parameters(), lr=lr, 
                  betas=(0.9, 0.98), eps=1e-06)

total_steps = len(dl) * epochs
warmup_steps = int(total_steps * warmup_ratio)
scheduler = get_linear_schedule_with_warmup(optimizer, warmup_steps, total_steps)
```

```python
from tqdm.auto import tqdm, trange
from torch.cuda.amp import GradScaler, autocast
import torch.nn as nn

from training_recorder import RunningLossRecorder
from jiwer import wer, compute_measures

import wandb
import os
```

```python
def compute_wer(preds, gts):
    incorrect = 0 
    total = 0 
    for pred, gt in zip(preds, gts): 
        measures = compute_measures(gt, pred) 
        incorrect += measures["substitutions"] + measures["deletions"] + measures["insertions"] 
        total += measures["substitutions"] + measures["deletions"] + measures["hits"] 
    return incorrect / total 
```

```python
def filter_tokens(a):
    ret = []
    start = False
    for x in a:
        if x in set([0, 1, 2]):
            if start:
                return ret
            else:
                start = True
        else:
            ret.append(x)
    return ret
```

```python
def mask_tokens(inputs, tokenizer, mlm_probability=0.15):
    mask_id = tokenizer.encode('<mask>', add_special_tokens=False).ids[0]
    vocab_size = tokenizer.get_vocab_size()
    
    """ Prepare masked tokens inputs/labels for masked language modeling: 80% MASK, 10% random, 10% original. """
    labels = inputs.clone()
    # We sample a few tokens in each sequence for masked-LM training (with probability args.mlm_probability defaults to 0.15 in Bert/RoBERTa)
    probability_matrix = torch.full(labels.shape, mlm_probability).to(device)
    #special_tokens_mask = [tokenizer.get_special_tokens_mask(val, already_has_special_tokens=True) for val in labels.tolist()]

    probability_matrix.masked_fill_(torch.eq(labels, 0), value=0.0)

    masked_indices = torch.bernoulli(probability_matrix).bool()
    labels[~masked_indices] = -100  # We only compute loss on masked tokens
 
    # 80% of the time, we replace masked input tokens with tokenizer.mask_token ([MASK])
    indices_replaced = torch.bernoulli(torch.full(labels.shape, 0.8)).to(device=device, dtype=torch.bool) & masked_indices
    inputs[indices_replaced] = mask_id

    # 10% of the time, we replace masked input tokens with random word
    indices_random = torch.bernoulli(torch.full(labels.shape, 0.5)).to(device=device, dtype=torch.bool) & masked_indices & ~indices_replaced
    random_words = torch.randint(vocab_size, labels.shape, device=device, dtype=torch.long)
    inputs[indices_random] = random_words[indices_random].to(device)

    # The rest of the time (10% of the time) we keep the masked input tokens unchanged
    return inputs, labels
```

```python
if not os.path.exists(f'{result_dir}/{model_save_name}'):
    os.mkdir(f'{result_dir}/{model_save_name}')
if use_wandb:
    wandb.init(project='ocr', entity='rolypolyvg295',
               tags=['training plm'], name=model_save_name)
    wandb.watch(model)
scaler = GradScaler(enabled=use_amp)
loss_recorder = RunningLossRecorder(log_step)
total_step = 0
for epoch in trange(epochs):
    model.train()
    print(f'Epoch: {epoch}')
    for step, batch in tqdm(enumerate(dl), total=len(dl)):
        optimizer.zero_grad()
        
        input_ids, attn_masks = [
            b.to(device) if torch.is_tensor(b) else b for b in batch
        ]
        
        with torch.no_grad():
            input_ids, labels = mask_tokens(input_ids, tokenizer)
            labels[labels == tokenizer.padding['pad_id']] = -100

        with autocast(enabled=use_amp):
            output = model(
                input_ids=input_ids,
                attention_mask=attn_masks,
                labels=labels,
            )
        
        loss = output.loss
        loss_recorder.add(loss.cpu().item())
        
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)
        scaler.step(optimizer)

        scale = scaler.get_scale()
        scaler.update()
        step_scheduler = scaler.get_scale() == scale

        if step_scheduler:
            scheduler.step()
            
        if step == len(dl) - 1 or (step != 0 and step % log_step == 0):
            logged_loss = loss_recorder.loss()[1]
#             print(f'Loss: {logged_loss}')
            if use_wandb:
                wandb.log({'loss':logged_loss, 'epoch':epoch + step/len(dl)}, step=total_step + step)
            
    total_step += step
#     print(f'Start Validation {epoch}')
    val_loss_recorder = RunningLossRecorder(len(val_dl))
    model.eval()
    gt = []
    prd = []
    
    for step, batch in tqdm(enumerate(val_dl), total=len(val_dl)):
        input_ids, attn_masks = [
            b.to(device) if torch.is_tensor(b) else b for b in batch
        ]
        
        with torch.no_grad():
            no_masked_labels = input_ids.detach().cpu().clone()
            input_ids, labels = mask_tokens(input_ids, tokenizer)
            labels[labels == tokenizer.padding['pad_id']] = -100

        with torch.no_grad():
            with autocast(enabled=use_amp):
                output = model(
                    input_ids=input_ids,
                    attention_mask=attn_masks,
                    labels=labels,
                )
        
        loss = output.loss
        val_loss_recorder.add(loss.cpu().item())

        pred = output['logits']
        pred = torch.max(pred, dim=-1)[1]
        
#         no_masked_labels[labels.eq(-100)] = -100
        list_labels = [filter_tokens(x) for x in no_masked_labels.tolist()]
        list_pred = [filter_tokens(x) for x in pred.tolist()]
        ground_truth = tokenizer.decode_batch(list_labels)
        ans = tokenizer.decode_batch(list_pred)
        
        gt.extend(ground_truth)
        prd.extend(ans)
    
    wer_error = compute_wer(prd, gt)
    correct = 0
    tot = len(gt)
    for i in range(tot):
        if gt[i] == prd[i]:
            correct += 1
    accr = correct / tot
    
    val_logged_loss = val_loss_recorder.loss()[1]
#     print(f'Val Loss: {val_logged_loss}')
#     print(f'Val Accr: {accr}')
#     print(f'Val  Wer: {wer_error}')
    if use_wandb:
        wandb.log({
            'val loss':val_logged_loss,
            'accr': accr,
            'wer': wer_error,
            'epoch':epoch + 1,
        }, step=total_step)
    print()
    
    if (epoch + 1) % save_epoch == 0:
        model.save_pretrained(f'{result_dir}/{model_save_name}/epoch_{epoch+1}')
        
if use_wandb:
    wandb.finish()
print('done')
```
